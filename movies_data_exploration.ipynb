# Handling messy data - converting "budget" to a numeric field

```python
movies_data["budget"]
```

```python
check = movies_data.budget.map(lambda val: str(val).strip("$").strip())
check
```

```python
#check.map(float) ## error because of messy data, not every value can be converted to a float (numeric) value
```

```python
def clean_currency(val):
    try:
        return float(val)
    except:
        return np.nan
```

```python
check = check.map(clean_currency)
```

```python
movies_data["budget"] = check
```

## Use similar logic to clean up the gross income fields

```python
movies_data["worlwide_gross_income"] = movies_data.worlwide_gross_income.map(lambda val: clean_currency(str(val).strip("$").strip()))
movies_data["usa_gross_income"] = movies_data.usa_gross_income.map(lambda val: clean_currency(str(val).strip("$").strip()))
```

```python
# movies_data.date_published.map(pd.Timestamp) # error 
```

```python
# create function to force clean date
def clean_date(s):
    try:
        return pd.Timestamp(s)
    except Exception:
        return np.nan
```

```python
movies_data["date_published"] = movies_data.date_published.map(clean_date)
```

```python
movies_data["date_published"].head()
```

```python
movies_data["release_month"] = movies_data.date_published.map(lambda dt: dt.month_name())
```

```python
movies_data.release_month[:10]
```

```python
movies_data["release_year"] = movies_data.date_published.map(lambda dt: dt.year)
```

```python
movies_data["release_day"] = movies_data.date_published.map(lambda dt: dt.day_name())
```

```python
movies_data.head()
```
